## Data Description and Preprocessing

For this project, you will explore data from the [National Health and Nutrition Examination Survey](https://www.kaggle.com/cdc/national-health-and-nutrition-examination-survey?select=questionnaire.csv). NHANES is a unique study that combines survey methodology with in-person medical examinations to create a dataset with demographic information, health indicators, and health outcomes.

We start by importing the data and doing some preliminary preprocessing for you. We import some libraries that will be helpful as well. 'SEQN' is the ID number for each respondent, and 'HSD010' will be our target outcome. [HSD010](https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/HSQ_H.htm#HSD010) asks for the respondent's self reported health condition, which can range from "excellent" to "poor."

```python
# Import libraries
#!pip3 install keras
#!pip install tensorflow
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
# Load nhanes data
nhanes = pd.read_csv('nhanes.csv')
# Get the ID numbers for each observation (seqn)
seqn = nhanes['SEQN']
# Get the target, "self-reported health condition," HSD010
hsd010 = nhanes['HSD010']
# Drop SEQN from the dataframe and then apply the standard scaler

nhanes_features = nhanes.drop(['SEQN', 'HSD010'], axis=1)
nhanes_scaled_features = pd.DataFrame(StandardScaler().fit_transform(nhanes_features),
                                       columns=nhanes_features.columns)
nhanes_scaled = pd.concat([nhanes_scaled_features, nhanes[['SEQN', 'HSD010']]], axis=1)
nhanes_scaled = nhanes_scaled.set_index('SEQN')
nhanes_scaled.head()
```

### Train PCA and Discuss Number of Components

```python
# Instantiate the PCA model
pca = PCA()
# Fit the model to your data
pca.fit(nhanes_scaled_features)

# Get the explained variance for each component
explained_variance = pca.explained_variance_

# Plot the explained variance for each component
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Variance Explained by Each Principal Component')
plt.show()

cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
cumulative_explained_variance = np.round(cumulative_explained_variance,3)
print(cumulative_explained_variance)
for i, explained_var in enumerate(cumulative_explained_variance):
    if explained_var > 0.85:
        print(f"Number of components before 0.85: {i + 1}")
        break

```

<span style="color:green; font-size: 20px; font-weight: bold"> Here's the scree plot for my PCA. I did a bit of reading online to figure out how to choose the number of principal components to use. I learned that if my sole intention of doing PCA is for data visualization, I should select 2 or 3 principal components for 2D or 3D data visualization. However, since we are trying to increase the interpretability of data while still preserving the maximum amount of information, our goal is to find the number of principal components which keeps as much of the variance in the original data as possible while meaningfully reducing dimensionality/complexity.</span>

<span style="color:green; font-size: 20px; font-weight: bold"> I included the cumulative explained variance below, and chose an arbitrary threshold of 85% variance. It cuts off at a nice round 100 principal components, as opposed to the larger 241 cut off at 145 components. Another factor to consider is computational efficiency and how much longer it would take for a neural network to use 100 vs. 145 vs. 213 components (213 because there are 28 components which explain 0 variance). For that reason, I decided to go with 100 components.</span>

### How many components will you use to fit a supervised learning model?

<span style="color:green; font-size: 20px; font-weight: bold"> I am going with 100 components to begin with. There are a lot of things to consider for fitting a supervised learning model. We want to keep enough components to capture a high percentage of the original data's variance. We also want to look at the performance of our supervised learning model on a validation set, and see how different numbers of components perform better or worse. Though I might answer 100 components here, I would also want to experiment with different numbers and observe how the model's performance changes. We should also consider computational efficiency. Having a smaller number of princiapl components improves computational efficiency. If the training time for 100 components is really high, I would be fine reducing the number of components. Overfitting is when the model captures noise/random variance in the training data rather than the underlying patterns of the relationship we are studying. </span>

### Choose a Clustering Algorithm

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Using the elbow method to determine the optimal number of clusters
inertia_values = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, 
                    random_state=144, 
                    max_iter=300,
                    n_init=10)
    kmeans.fit(nhanes_scaled_features)
    inertia_values.append(kmeans.inertia_)
```

<span style="color:green; font-size: 20px; font-weight: bold"> I am going with K-means clustering as my clustering algorithm. K-means clustering partitions a dataset into a predefined number of clusters (k). It groups data points into clusters by minimizing the total squared distance between each point and the "centroid" of its assigned cluster. It iteratively updates so that the centroids are recalculated based on the means of all data points assigned to their respective clusters. The new centroids become the center of their clusters.</span>

### Cluster nhanes

```python
# Plotting the elbow method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 20), inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances Between Data Point and Center of Cluster')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

inertia = []
K_range = range(2, 11)

silhouette_scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(nhanes_scaled_features)
    silhouette_scores.append(silhouette_score(nhanes_scaled_features, kmeans.labels_))

plt.plot(K_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()
```

```python
optimal_k = 4

kmeans = KMeans(n_clusters=optimal_k, random_state=169, n_init=10)
nhanes_clusters = kmeans.fit_predict(nhanes_scaled_features)
```

<span style="color:green; font-size: 20px; font-weight: bold"> I'm using k=4. I used the silhouette method and the elbow method. The silhouette score measures how similar an object is to its own cluster, compared to other clusters, for any given k value. The suggests to choose the value of k with the highest silhouette score. The elbow method calculates the sum of squared distances of each point to its assigned cluster centroid and plots these values for each k value. The "elbow" point in the plot is often where the optimal k is, and this method more explicitly strikes a balance between model simplicity and explanatory power.</span>

<span style="color:green; font-size: 20px; font-weight: bold"> The silhouette score was highest for k=2 and it is a bit unclear where the elbow is in the elbow method graph, but I eyeballed roughly k=4. Either way, these two methods confirmed that the k-value in this case would be quite low, between 2-4. I was originally going to go with k=3 because it happened to be right in the middle of these two numbers. However, after using both k=3 and k=4, I noticed that k=4 slightly improved model performance and also produced a more useful data visualization with the BMI vs. Income Poverty Ratio Plot.</span>

### Neural Network on Original Features

<span style="color:orange; font-size: 20px; font-weight: bold"> First, I ran a Neural Network on Original Features for HSD010. I know the assignment asked us to choose either one or the other, but I was curious so I ran both. If you can only grade one, look at the binary models in purple. </span> 

```python
# load libraries
# -----------
import keras
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, LeakyReLU

# convert hsd010 from text back to numerical values
nhanes_scaled['HSD010'] = nhanes_scaled['HSD010'].replace(
    ['excellent', 'very good', 'good', 'fair', 'poor'], [0, 1, 2, 3, 4])
num_classes = len(nhanes_scaled['HSD010'].unique())

# partition data
y = nhanes_scaled['HSD010']
X = nhanes_scaled_features

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = .25, 
                                                    random_state = 30)

# convert class vectors to binary class matrices
y_train = to_categorical(y_train, num_classes=num_classes)
y_test = to_categorical(y_test, num_classes=num_classes)
```

```python
# create neural network model
model = Sequential()
model.add(Dense(144, input_dim=X_train.shape[1], kernel_initializer='normal', activation='relu'))
model.add(Dense(51, kernel_initializer='normal', activation='relu'))

# add any additional layers you wish here
model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
original_model = model.fit(X_train,                  # specify the x data
                           y_train,                  # specify the y data
                           validation_data=(X_test,  # specify the validation data
                                            y_test), 
                           epochs=20,                # number of iterations back and forth through the network
                           batch_size=200,            # number of inputs per batch
                           verbose=2);               # print as it goes (similar to folds in cross validation)

# Print the final accuracy values in bold
print(f"\n\033[1mFinal Training Accuracy (NN Model on Original Features, Categorical Outcome): {original_model.history['accuracy'][-1]}\033[0m")
print(f"\033[1mFinal Validation Accuracy (NN Model on Original Features, Categorical Outcome): {original_model.history['val_accuracy'][-1]}\033[0m")
```

<span style="color:purple; font-size: 20px; font-weight: bold"> Next, I also ran a Neural Network on Original Features for HSD010_binary. ## Discussion Questions

1. In your own words, what is the difference between PCA and clustering?

<span style="color:green; font-size: 20px; font-weight: bold"> PCA is a dimensionality reduction technique which transforms high-dimensional data into a lower-dimensional representation. PCA identifies the most significant features, or principal components, which capture the most variance in the data. Clustering is a grouping technique which categorizes data points into clusters based on their similarity (measured by distance to a given centroid) in an iterative process. PCA focuses on reducing feature dimensions and retaining as much information as possible, while clustering focuses on grouping data points together based on their similarity without reducing dimensionality.</span>

2. Did you notice any advantages to combining PCA and clustering? If so, what do you think they were? If not, why do you think you didn't see any gains from this combination?

<span style="color:green; font-size: 20px; font-weight: bold"> Yes, I did notice advantages to combining PCA and clustering. The neural network which used both prinicipal components and cluster membership performed better in accuracy and validation accuracy than the neural network which only used PCA. PCA and clustering provide complementary information: PCA captures essential features by reducing dimensionality while clustering identifies similar groupings in the data. When used together, these methods create a more robust feature representation: PCA is good at finding macro-level "global" patterns, while clustering is good at finding out about "local" structures. Combining them together may help the neural network capture more complex relationships.</span>

3. How can unsupervised techniques help with downstream supervised learning tasks when working with "big data?"

<span style="color:green; font-size: 20px; font-weight: bold">I can think of at least three ways which unsupervised machine learning tools can help support supervised learning. First, unsupervised methods are useful for preprocessing data. In class, we discussed the usefulness of unsupervised learning for scaling, normalization, and estimating missing values. Second, as we saw in this project, unsupervised methods like PCA can help draw out the most important features from high-dimensional data, in dimensionality reduction. Dimensionality reduction can help make data more interpretable, or easy to understand. It also is useful in ensuring computational efficiency; if a dataset has a huge number of features, it may take too long to run supervised learning methods on the entire thing without reducing dimensions. Third, we can use clustering algorithms to reveal patterns within the data. Understanding clusters or local structures can then guide our feature selection process for supervised learning, saving time and potentially identifying insights which would otherwise have gone unnoticed.</span>
